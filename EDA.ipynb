```python
df = df.reset_index(drop=True)
```

```python
new_df = df.drop(columns=['adult','imdb_id','belongs_to_collection','budget','homepage','status','original_title','video','production_companies','production_countries','revenue','spoken_languages','tagline','original_language'])
```

```python
new_df.head()
```

```python
import ast

def extract_tag_names(tag_string):
    try:
        tag_list = ast.literal_eval(tag_string)  # safely parse the string
        return [tag['name'] for tag in tag_list if 'name' in tag]
    except Exception:
        return []  # in case of bad formatting or NaNs
new_df['genres'] = new_df['genres'].apply(extract_tag_names)
```

```python
temp_df = new_df[['popularity','runtime','vote_average']]
```

```python
temp_df['popularity']
```

```python

# Find non-float values
non_floats = temp_df[~temp_df['popularity'].apply(lambda x: isinstance(x, float))]

# Display suspicious rows
print(non_floats)
```

```python
def is_convertible_to_float(x):
    try:
        float(x)
        return True
    except (ValueError, TypeError):
        return False

# Find rows where the value cannot be converted to float
non_convertible = temp_df[~temp_df['popularity'].apply(is_convertible_to_float)]

print(non_convertible)
```
